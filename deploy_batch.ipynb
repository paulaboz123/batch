# Azure ML Batch Endpoint (AAD) – Deploy + Smoke Test
This notebook deploys a batch endpoint + 3 deployments (de/pl/en) and submits a smoke-test invoke.


```python
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential
from azure.ai.ml.entities import BatchEndpoint, BatchDeployment
```

```python
SUBSCRIPTION = '<subscription>'
RESOURCE_GROUP = '<resource_group>'
WORKSPACE = '<workspace>'
ml_client = MLClient(DefaultAzureCredential(), SUBSCRIPTION, RESOURCE_GROUP, WORKSPACE)
```

```python
endpoint = BatchEndpoint.load('deployment/endpoint-dev.yml')
ml_client.batch_endpoints.begin_create_or_update(endpoint).result()
```

```python
for f in ['deployment/de-model.yml','deployment/pl-model.yml','deployment/en-model.yml']:
    dep = BatchDeployment.load(f)
    ml_client.batch_deployments.begin_create_or_update(dep).result()
```

## Smoke test (CLI recommended)
Batch invoke is asynchronous. In CI it's easiest to use `az ml batch-endpoint invoke`.
